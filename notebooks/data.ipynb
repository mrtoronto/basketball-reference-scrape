```python
import scraper
print('scraper version loaded. Base URL:', scraper.BASE_URL)
```

```python
# Search for a player by name
results = scraper.search_player_ids("LeBron James")
ids = [r['id'] for r in results]

results = scraper.search_player_ids("Michael Jordan")
ids += [r['id'] for r in results]

results = scraper.search_player_ids("Kobe Bryant")
ids += [r['id'] for r in results]

ids = list(set(ids))
ids
```

```python
seasons = [2023, 2024]

combined_data = {}

for s in seasons:
    combined_data[s] = {}
    for pid in ids:
        combined_data[s][pid] = []
        try:
            rows = scraper.scrape_player_game_logs(pid, s, last_n=15)
        except:
            print(f"Error scraping data for {pid} in {s}")
            continue
        for row in rows:
            row2 = dict(row)
            row2["PlayerID"] = pid
            combined_data[s][pid].append(row2)

```

```python
combined_data
```

```python
averages = {}


for s in seasons:
    averages[s] = {}
    for pid in ids:
        averages[s][pid] = {}
        if len(combined_data[s][pid]) == 0:
            continue
        games = combined_data[s][pid]
        stat_values = list(games[0].keys())
        new_stat_values = [k + "_avg" for k in stat_values]
        for stat in stat_values:
            total = 0
            for game in games:
                try:
                    total += int(game[stat])
                except:
                    continue
            averages[s][pid][stat + "_avg"] = total / len(games)

```
